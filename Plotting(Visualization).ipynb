# Hexagonal Bin Plot: hexbin()

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=["a", "b"])
df['b'] = df['b'] + np.arange(1000)
df
```

```python
df.plot.hexbin(x="a", y="b", gridsize=20)  # gridsize -- controls the number of hexagons in the x-direction,larger size means more smaller bins
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns = ['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df['z'] = np.random.uniform(0,3,1000)
```

```python
df.plot.hexbin(x="a", y="b", C="z",reduce_C_function=np.max, gridsize=25)  # Doubtful(reduce_C_function)
```

# 5.Lag plot
used to check if a data set is random

```python
from pandas.plotting import lag_plot
plt.figure()
spacing = np.linspace(-99 * np.pi, 99 * np.pi, num=1000) # Gives spaced number over a given interval
data = pd.Series(0.1 * np.random.rand(1000) + 0.9 * np.sin(spacing))
data
```

```python
lag_plot(data)
```

```python
df = pd.DataFrame(np.random.rand(1000))  # Lag Plot for Random Numbers
lag_plot(df) 
```

```python
# Plotting on a secondary y-axis
df = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range("1/1/2000", periods=1000), columns=list("ABCD")).cumsum()
df["A"].plot()
df["B"].plot(secondary_y=True, style="g")
```

```python
# If you have more than one plot that needs to be suppressed
with pd.plotting.plot_params.use("x_compat", True):
    df["A"].plot(color="r")
    df["B"].plot(color="g")
    df["C"].plot(color="b")
```

```python
ix3 = pd.MultiIndex.from_arrays(
       [
           ["a", "a", "a", "a", "b", "b", "b", "b"],
           ["foo", "foo", "bar", "bar", "foo", "foo", "bar", "bar"],
       ],
       names=["letter", "word"],
)

df3 = pd.DataFrame(
    {
        "data1": [3, 2, 4, 3, 2, 4, 3, 2],
        "data2": [6, 5, 7, 5, 4, 5, 6, 5],
    },
    index=ix3,
)
ix3
df3
```

```python
# Group by index labels 
gp3 = df3.groupby(level=("letter", "word"))
```

```python
means = gp3.mean()
means
```

```python
errors = gp3.std()
errors
```

```python
fig, ax = plt.subplots()
means.plot.bar(yerr=errors, ax=ax, capsize=4, rot=0)
```

# Plotting with Tables

```python
df = pd.DataFrame(np.random.rand(5, 3), columns=["a", "b", "c"])
df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
ax.xaxis.tick_top()  # Display x-axis ticks on top.
df.plot(table=True, ax=ax)
# The simple way to draw a table is to specify table=True.
```

```python
from pandas.plotting import table
fig, ax = plt.subplots(1, 1)
table(ax, np.round(df.describe(), 2), loc="upper right", colWidths=[0.2, 0.2, 0.2])
df.plot(ax=ax, ylim=(0, 2), legend=None)  # ylim refers to y axis's max limit
```
